```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-0.5, 0.5, 6)
y = np.linspace(-0.5, 0.5, 6)
```

```python
x, x.shape
```

```python
for xt in x:
    plt.plot([xt,xt], [y[0], y[-1]], 'k-')
    for yt in y:
        plt.plot([x[0], x[-1]], [yt,yt], 'k-')
plt.xlim((-0.6, 0.6))
plt.ylim((-0.6, 0.6))
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
N = 500
xp = (np.random.standard_normal(N))
yp = (np.random.standard_normal(N))

xmax, xmin = np.max(xp), np.min(xp)
ymax, ymin = np.max(yp), np.min(yp)

xp = 0.8 * xp / (xmax-xmin)
yp = 0.8 * yp / (ymax-ymin)
```

```python
np.min(xp), np.max(xp), np.min(yp), np.max(yp)
```

```python
plt.plot(xp, yp, 'o')
```

```python
import h5py

d = h5py.File('data.h5', 'w')
d['x'] = xp[:]
d['y'] = yp[:]
d.close()
```

```python
# test file

with h5py.File('data.h5', 'r') as d:
    x = d['x'][:]
    y = d['y'][:]

plt.plot(x,y, 'ko')
```
